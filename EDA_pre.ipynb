```python
#Se importan las librerias
import pandas as pd
import numpy as np
import ast
import json
import pickle
```

```python
#Se cargan los archivos de la carpeta metadata-sitios y se compilan en un solo DataFrame

for i in range(1,12):

    metadata_sitios=[]

    archivo=[]

    with open(f"/content/drive/My Drive/Google Maps/metadata-sitios/{i}.json","r", encoding= 'utf-8') as filejson:


            # Procesa cada línea del archivo como un objeto JSON
            for linea in filejson:

                    # Intenta cargar la línea como un objeto JSON
                    objeto = json.loads(linea)

                    archivo.append(objeto)

    metadata_sitios+=archivo

metadata_sitios=pd.DataFrame(metadata_sitios)
```
